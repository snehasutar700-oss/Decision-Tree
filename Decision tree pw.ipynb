```python
#Q16. Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#Q21. Write a Python program to train a Decision Tree Classifier and visualize the tree using graphviz

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

full = DecisionTreeClassifier()
full.fit(X_train, y_train)

shallow = DecisionTreeClassifier(max_depth=3)
shallow.fit(X_train, y_train)

print("Full Tree:", accuracy_score(y_test, full.predict(X_test)))
print("Depth=3 Tree:", accuracy_score(y_test, shallow.predict(X_test)))
```

```python
#Q22. Write a Python program to train a Decision Tree Classifier with a maximum depth of 3 and compare its accuracy with a fully grown tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

default = DecisionTreeClassifier()
default.fit(X_train, y_train)

model = DecisionTreeClassifier(min_samples_split=5)
model.fit(X_train, y_train)

print("Default:", accuracy_score(y_test, default.predict(X_test)))
print("min_samples_split=5:", accuracy_score(y_test, model.predict(X_test)))
```

```python
#Q23. Write a Python program to train a Decision Tree Classifier using min_samples_split=5 and compare its accuracy with a default tree*

from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

model1 = DecisionTreeClassifier()
model1.fit(X_train, y_train)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_scaled, y)

model2 = DecisionTreeClassifier()
model2.fit(X_train_s, y_train_s)

print("Without Scaling:", accuracy_score(y_test, model1.predict(X_test)))
print("With Scaling:", accuracy_score(y_test_s, model2.predict(X_test_s)))
```

```python
#Q26. Write a Python program to train a Decision Tree Regressor with max_depth=5 and compare its performance with an unrestricted tree*

from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

data = fetch_california_housing()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

full = DecisionTreeRegressor()
full.fit(X_train, y_train)

limited = DecisionTreeRegressor(max_depth=5)
limited.fit(X_train, y_train)

print("Full Tree MSE:", mean_squared_error(y_test, full.predict(X_test)))
print("Depth=5 MSE:", mean_squared_error(y_test, limited.predict(X_test)))
```

```python
#Q27.  Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = DecisionTreeClassifier(ccp_alpha=0.01)
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
#Q30. Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values

for max_depth and min_samples_split.

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_iris

data = load_iris()
X, y = data.data, data.target

params = {
    "max_depth": [2, 3, 4, 5],
    "min_samples_split": [2, 5, 10]
}

grid = GridSearchCV(DecisionTreeClassifier(), params)
grid.fit(X, y)

print("Best Parameters:", grid.best_params_)
```
